## Data Cleaning

```python
## check if all values in number of reviews
df['Reviews'].unique()
```

```python
df['Reviews'].str.isnumeric().sum()
```

```python
df['Reviews'].astype(int)
```

```python
df[~df['Reviews'].str.isnumeric()]
```

```python
df_copy=df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df['Reviews'].str.isnumeric()]
```

```python
## Convert Review DataType to int
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device', np.nan)
df_copy['Size']=df_copy['Size'].astype('float')
```

```python
df_copy['Size']
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')
```

```python
df_copy.info()
```

```python
## Handle last update feature
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.head()
```

```python
df_copy.to_csv("google_cleaned.csv")
```

## Explore Data

```python
numerical_features=[feature for feature in df_copy.columns if df_copy[feature].dtype!='O']
categorical_features=[feature for feature in df_copy.columns if df_copy[feature].dtype=='O']
print(f'We have {len(numerical_features)} numerical features: {numerical_features}')
print("\nWe have {} categorical features: {}".format(len(categorical_features), categorical_features))
```

```python
df_copy['Category'].dtype != 'O'
```

```python
## Proportion of count data on categorical columns
for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print("----------------------")
```

```python
## Proportion of count data on categorical column
plt.figure(figsize=(15,15))
plt.suptitle("Univariant Analysis of Numerical Features",fontsize=20,fontweight='bold',alpha=0.8,y=1.)
for i in range(0,len(numerical_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_copy[numerical_features[i]],fill=True,color='r')
    plt.xlabel(numerical_features[i])
    plt.tight_layout()
    if i==len(numerical_features)-1:
        plt.show()
```

## Observation
- Rating and Year is left skewed while Reviews, Size, Installs and Price are right skewed

```python
df_copy.columns
```

```python
# Categorical Columns
plt.figure(figsize=(20,15))
plt.suptitle("Univariant Analysis of Categorical Features",fontsize=20,fontweight='bold',alpha=0.8,y=1.)
category=['Type','Content Rating']
for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df_copy[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
    if i==len(category)-1:
        plt.show()
```

```python
## Top 10 App Categories
category=pd.DataFrame(df_copy['Category'].value_counts())
category.rename(columns={'Category':'Count'},inplace=True)
```

```python
plt.Figure(figsize=(15,6))
sns.barplot(x=category.index[:10],y='count',data=category[:10],palette='hls')
plt.title("Top 10 App Categories")
plt.xticks(rotation=90)
plt.show()
```

```python
app_install=pd.DataFrame(df_copy.groupby('Category')['Installs'].sum())
app_install.rename(columns={"Category": "Installs"},inplace=True)
app_install=app_install.sort_values(by="Installs",ascending=False)

```

```python
plt.Figure(figsize=(15,6))
sns.barplot(x=app_install.index[:10],y='Installs',data=app_install[:10],palette='hls')
plt.title("Top 10 App Installs")
plt.xticks(rotation=90)
plt.show()
```
