```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import MinMaxScaler 
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("../data/income.csv")
df.head()
```

```python
plt.scatter(df['Age'], df['Income($)'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'], color='green')
plt.scatter(df2.Age,df2['Income($)'], color='red')
plt.scatter(df3.Age,df3['Income($)'], color='black')
# plt.scatter(km.cluster_centers_[:,0], km.cluster_centers[:,1],color="purple", marker='*', label='centroid')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
df.head()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])  # [['Income($)']] returns a 2D DataFrame
df['Income($)'] = scaler.transform(df[['Income($)']])  # Use double brackets here as well

scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
df
```

```python
plt.scatter(df['Age'], df['Income($)'])
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'], color='green')
plt.scatter(df2.Age,df2['Income($)'], color='red')
plt.scatter(df3.Age,df3['Income($)'], color='black')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],color="purple", marker='*', label='centroid')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
k_rng = range(1,11)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age', 'Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_rng, sse)
plt.xlabel('K')
plt.ylabel('Sum of squared error')
```
